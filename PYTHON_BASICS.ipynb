### Bubble sort

Bubble sort : [4 3 2 1] 

Wost case time complexity : O(n*(n-1)) ~ O(n^2) 
Space complexity : O(1)

- Pass 1 : Compare 1st, 2nd element and exchange if 1st element > 2nd element and keep going
  - 1 => 3 4 2 1
  - 2 => 3 2 4 1
  - 3 => 3 2 1 4
  - Result : 3 2 1 4 (One element is sorted)
- Pass 2 : 2 1 3 4
- Pass 3 : 1 2 3 4'''

```python
def bubble_sort(lst):

  # Boundary case
  if len(lst) <= 1:
    return lst

  for i in range(0, len(lst)):
    
    # ith pass sorting
    for ind in range(0, len(lst)-1):
      if lst[ind] > lst[ind+1]:
        tmp = lst[ind]
        lst[ind] = lst[ind+1]
        lst[ind+1] = tmp

    print(lst) # By now one pass is sorted

  return lst


a = [4, 2, 1, 3]
print(bubble_sort(a))
```

### Insertion sort

Insertion sort : [4 3 2 1]

Ref: Insertion sort working : https://www.google.com/search?q=insertion+sort+working+youtube&oq=insertion+sort+working+youtu&aqs=chrome.1.69i57j33i10i160l3.9421j0j7&sourceid=chrome&ie=UTF-8#kpvalbx=_gDPGY4XSJdyUseMPk6CKmAY_27

- Pass 1 : Compare first two elements and sort :3421
  - Now sorted part is first two elements [3 4]
  - Next take next element from unsorted array [2 1] i.e. 2 and insert in sorted array at the correct position
- Pass 2 : 2341
  - Sorted part [2 3 4] Unsorted [1]
- Pass 3 : 1234 

Wost case time complexity : O(n*(n-1)/2) ~ O(n^2) 
Space complexity : O(1)



```python
def insertion_sort(InputList):

  # Boundary case
  if len(InputList) <= 1:
    return 

  # Unsorted array starts from index 1
  for unsort_idx in range(1, len(InputList)):
    # print(unsort_idx)

    # Sorted array index starts from 0
    idx = 0

    while idx < unsort_idx:

      # If unsorted index element < any element in sorted array exchange with unsorted element
      if InputList[idx] > InputList[unsort_idx]:
        tmp = InputList[idx]
        InputList[idx] = InputList[unsort_idx]
        InputList[unsort_idx] = tmp
        print(InputList)

      idx += 1



list = [4, 2, 1, 3]
insertion_sort(list)
print(list)
```

### Selection sort

Selection sort : [4 3 2 1]

Ref: Selection sort working : https://www.google.com/search?q=selection+sort+working&sxsrf=AJOqlzUae4BUtuxENnX00K2ySQvYD5E8oQ:1673936639178&source=lnms&tbm=vid&sa=X&ved=2ahUKEwj7tIT2-838AhVPcGwGHa4iDxMQ_AUoAnoECAEQBA&biw=1848&bih=949&dpr=1#fpstate=ive&vld=cid:4fa0b447,vid:g-PGLbMth_g

- Pass 1 : Compare first element with remaining elements, find the minimum and swap with first index : 1 3 2 4
  - Now sorted part is first two elements [1]
  - Next take min element from unsorted array [3 2 4] i.e. 2 and insert in sorted array at the correct position
- Pass 2 : 1 2 3 4
  - Sorted part [1 2] Unsorted [3 4]
- Pass 3 : 1 2 3 4
  - Sorted part [1 2 3] Unsorted [4]

Wost case time complexity : O(n*(n-1)/2) ~ O(n^2) 
Space complexity : O(1)


```python
def selection_sort(InputList):

  # Boundary case
  if len(InputList) <= 1:
    return 

  for idx in range(0, len(InputList)):
    

    # Find the minimum of unsorted array
    i, min_idx = idx, idx

    while i < len(InputList):
      if InputList[min_idx] > InputList[i]:
        min_idx = i
      i += 1

    # Swap with sorted array index and the minimum element
    tmp = InputList[min_idx]
    InputList[min_idx] = InputList[idx]
    InputList[idx] = tmp

    print(InputList)


list = [54, 31, 12, 10 ]
selection_sort(list)
print(list)
```
